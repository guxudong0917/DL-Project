# 神经网络
基于pytorch框架，构建一个神经网络已经变得非常容易，就像搭积木一样。网络继承nn.Module类，需要重写初始化函数和前向传播函数

下面写一个最简单的神经网络，用来二分类

```python
import torch
import torch.nn as nn
class Net(nn.Module):
    def __init__(self,input_size,hidden_size,output_size):
        #用完成父类的初始化，不可以省略
        super(Net, self).__init__()
        self.layer1=nn.Linear(input_size,hidden_size)
        self.layer2=nn.Linear(hidden_size,output_size)
        self.relu=nn.ReLU()
        self.sigmoid=nn.Sigmoid()

    def forward(self,x):
        h=self.relu(self.layer1(x))
        result=self.sigmoid(self.layer2(h))
        return result
```

nn(Nerual network)就像一个积木库，里面包含不同的神经网络层级，我们就选择权重层来搭建一个新的网络，然后在forward阶段一层一层的传播即可。

### nn.Linear(input,output)
`layer=nn.Linear(10,3)`这一步设定了一个size为[10，3]的权重矩阵，与输入[n,10]相乘后,会有[n,10][10,3]=[n,3]
### nn.Relu()
这个函数是一个激活函数，一般放在线性层后面，用来引入非线性。
### nn.sigmoid
这个函数一般在二分类的时候用，将输出转化为概率，越靠近0越可能是类别0，越靠近1越可能是类别1
### nn.softmax
多分类时用，计算是每一类的概率，和为1
